```python
import numpy as np

# Create a NumPy array with values from 10 to 49 (inclusive)
array = np.arange(10, 50)
print("Original array:")
print(array)
```

```python
# Create two 3×3 matrices: One filled with all 5s , One with values from 1 to 9
#● Add the two matrices
#● Multiply them element-wise
#● Then find the transpose of the result
```

```python
# Create a 3x3 matrix filled with all 5s
matrix_fives = np.full((3, 3), 5)
print("Matrix 1 (all 5s):\n", matrix_fives)
```

```python
# Create a 3x3 matrix with values from 1 to 9
matrix_one_to_nine = np.arange(1, 10).reshape(3, 3)
print("\nMatrix 2 (1 to 9):\n", matrix_one_to_nine)
```

```python
# Add the two matrices
matrix_sum = matrix_fives + matrix_one_to_nine
print("\nSum of the  matrices:\n", matrix_sum)
```

```python
# Multiply them element-wise
matrix_product = matrix_fives * matrix_one_to_nine
print("\n product of the two matrices:\n", matrix_product)
```

```python
# Find the transpose of the element-wise product
matrix_product_transpose = matrix_product.T
print("\nTranspose:\n", matrix_product_transpose)
```
