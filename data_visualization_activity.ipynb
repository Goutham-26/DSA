```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds['cut'].value_counts()
```

```python
sns.scatterplot(data=diamonds,x='carat',y='price',hue='clarity')
```

```python
sns.countplot(data=diamonds,x='cut')
```

```python
sns.histplot(data=diamonds,x='price',kde=True)
```

```python
sns.boxplot(data=diamonds,x='cut',y='price')
```

```python
sns.violinplot(data=diamonds,x='cut',y='price')
```

```python
sns.violinplot(data=diamonds, x='cut', y='price',palette='magma',hue='cut')
```

```python
sns.stripplot(data=diamonds,x='cut',y='price')
```

```python
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', bins=30)
```

```python
sns.histplot(data=diamonds, x='price', hue='cut',multiple='stack',element='step', bins=50)
```

```python
corr=diamonds.corr(numeric_only=True)
corr
```

```python

sns.heatmap(corr,annot=True,cmap='Purples')
```

```python
sns.barplot(data=diamonds,x='cut',y='price',color='blue')
```
